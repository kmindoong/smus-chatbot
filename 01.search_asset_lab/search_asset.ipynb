### ë¨¼ì € ë‹¨ìˆœíˆ SMUS Domain ë‚´ ì „ì²´ Assetì„ ì¡°íšŒ

- Searchë©”ì„œë“œëŠ” searchScope ê¸°ì¤€ìœ¼ë¡œ ì¡°íšŒê°€ ê°€ëŠ¥í•˜ë©°, ì•„ë˜ì™€ ê°™ì´ 4ê°€ì§€ë¥¼ ì§€ì›
- searchScope: "ASSET" || "GLOSSARY" || "GLOSSARY_TERM" || "DATA_PRODUCT"

````python
#====ì „ì²´ Asset ì¡°íšŒ====
import boto3

dz = boto3.client("datazone", region_name="ap-northeast-2")
domain_id = "dzd-cjvglgj4d43fmg" # ë„ë©”ì¸ ID ì…ë ¥

# 1ï¸âƒ£ ëª¨ë“  í”„ë¡œì íŠ¸ ì¡°íšŒ
projects_resp = dz.list_projects(domainIdentifier=domain_id)
project_ids = [p["id"] for p in projects_resp.get("items", [])]

all_assets = []

# 2ï¸âƒ£ ê° í”„ë¡œì íŠ¸ë³„ Asset ì¡°íšŒ
for pid in project_ids:
    next_token = None
    while True:
        params = {
            "domainIdentifier": domain_id,
            "owningProjectIdentifier": pid,
            "searchScope": "ASSET"
        }
        if next_token:
            params["nextToken"] = next_token

        resp = dz.search(**params)
        for item in resp.get("items", []):
            asset_item = item.get("assetItem", {})
            all_assets.append({
                "id": asset_item.get("identifier"),
                "name": asset_item.get("name"),
                "type": asset_item.get("typeIdentifier"),
                "project": asset_item.get("owningProjectId"),
            })

        next_token = resp.get("nextToken")
        if not next_token:
            break

print(f"âœ… Found {len(all_assets)} assets in domain {domain_id}")
for a in all_assets:
    print(f"{a['id']} | {a['name']} | {a['type']} | {a['project']}")
````

### Datazone Clientì˜ Searchì˜ ê²½ìš° ëª¨ë“  íƒ€ì…ì˜ Asset ì¤‘ typeIdentifierë¥¼ ì§€ì •í•˜ì—¬ ì¡°íšŒë„ ê°€ëŠ¥
### ì•„ë˜ ì˜ˆì‹œì—ì„œëŠ” ASSET ì¤‘ GlueTableAssetTypeë§Œ ì¡°íšŒí•˜ê³  Project Nameë„ ê°™ì´ ì¡°íšŒ

````python
import boto3

dz = boto3.client("datazone", region_name="ap-northeast-2")
domain_id = "dzd-cjvglgj4d43fmg"

# 1ï¸âƒ£ ë„ë©”ì¸ ë‚´ ëª¨ë“  í”„ë¡œì íŠ¸ ì¡°íšŒ
projects_resp = dz.list_projects(domainIdentifier=domain_id)
projects = {p["id"]: p["name"] for p in projects_resp.get("items", [])}  # id â†’ name mapping

# 2ï¸âƒ£ ëª¨ë“  í”„ë¡œì íŠ¸ë³„ Asset ì¡°íšŒ (GlueTableAssetTypeë§Œ)
glue_assets = []

for project_id, project_name in projects.items():
    next_token = None
    while True:
        params = {
            "domainIdentifier": domain_id,
            "owningProjectIdentifier": project_id,
            "searchScope": "ASSET"
        }
        if next_token:
            params["nextToken"] = next_token

        resp = dz.search(**params)

        for item in resp.get("items", []):
            asset_item = item.get("assetItem", {})

            if asset_item.get("typeIdentifier") == "amazon.datazone.GlueTableAssetType":
                glue_assets.append({
                    "id": asset_item.get("identifier"),
                    "name": asset_item.get("name"),
                    "type": asset_item.get("typeIdentifier"),
                    "project_id": project_id,
                    "project_name": project_name,
                })

        next_token = resp.get("nextToken")
        if not next_token:
            break

# 3ï¸âƒ£ ê²°ê³¼ ì¶œë ¥
print(f"âœ… Found {len(glue_assets)} GlueTableAssetType assets in domain {domain_id}\n")
for a in glue_assets:
    print(f"{a['id']} | {a['name']} | {a['type']} | {a['project_name']} ({a['project_id']})")
````

### ì•„ë˜ ì˜ˆì‹œëŠ” íŠ¹ì • í”„ë¡œì íŠ¸ ë‚´ì˜ ì „ì²´ Asset ì¡°íšŒ í›„ get_assetë©”ì„œë“œë¡œ í•´ë‹¹ Assetì— ëŒ€í•œ ë©”íƒ€ë°ì´í„° ê²°ê³¼ ì¶œë ¥
- get_asset ë©”ì„œë“œ ì •ë³´
- https://boto3.amazonaws.com/v1/documentation/api/latest/reference/services/datazone/client/get_asset.html

````python
import boto3
import json

region = "ap-northeast-2"
dz = boto3.client("datazone", region_name=region)

domain_id = "dzd-cjvglgj4d43fmg"
project_id = "bknisgxzopjuhk"

# 1ï¸âƒ£ Asset ê²€ìƒ‰
resp = dz.search(
    domainIdentifier=domain_id,
    owningProjectIdentifier=project_id,
    searchScope="ASSET"
)

assets = resp.get("items", [])
print(f"ğŸ” Found {len(assets)} assets\n")

if not assets:
    raise Exception("No assets found")

# 2ï¸âƒ£ ì „ì²´ Asset ìˆœíšŒ
for idx, asset in enumerate(assets, 1):
    asset_id = asset["assetItem"]["identifier"]
    asset_name = asset["assetItem"].get("name")
    print(f"=== Asset {idx} ===")
    print(f"Identifier: {asset_id}")
    print(f"Name: {asset_name}")

    # 3ï¸âƒ£ ë©”íƒ€ë°ì´í„° ì¡°íšŒ
    asset_resp = dz.get_asset(
        domainIdentifier=domain_id,
        identifier=asset_id
    )
    print("Metadata:")
    print(json.dumps(asset_resp, default=str, indent=2))

    # 4ï¸âƒ£ ìŠ¤í‚¤ë§ˆ ì •ë³´ê°€ ìˆë‹¤ë©´ ì¶œë ¥
    schema = asset_resp.get("metadata", {}).get("schema", None)
    if schema:
        print("Schema:")
        print(json.dumps(schema, default=str, indent=2))
    else:
        print("Schema: None")
    
    print("\n")  # êµ¬ë¶„
````

### ì•„ë˜ ì˜ˆì‹œëŠ” Asset ê²°ê³¼ë¥¼ JSON íŒŒì¼ë¡œ ì €ì¥

````python
import boto3
import json

region = "ap-northeast-2"
dz = boto3.client("datazone", region_name=region)

domain_id = "dzd-cjvglgj4d43fmg"
project_id = "bknisgxzopjuhk"

# 1ï¸âƒ£ Asset ê²€ìƒ‰
resp = dz.search(
    domainIdentifier=domain_id,
    owningProjectIdentifier=project_id,
    searchScope="ASSET"
)

assets = resp.get("items", [])
print(f"ğŸ” Found {len(assets)} assets\n")

if not assets:
    raise Exception("No assets found")

all_assets_data = []

for asset in assets:
    asset_id = asset["assetItem"]["identifier"]
    asset_name = asset["assetItem"].get("name")

    # 2ï¸âƒ£ Asset ìƒì„¸ ì¡°íšŒ
    asset_resp = dz.get_asset(
        domainIdentifier=domain_id,
        identifier=asset_id
    )

    # 3ï¸âƒ£ ì „ì²´ êµ¬ì¡° í™•ì¸ìš©
    # print(json.dumps(asset_resp, default=str, indent=2))

    # 4ï¸âƒ£ ì•ˆì „í•˜ê²Œ í•„ë“œ ì¶”ì¶œ
    # ì‹¤ì œ Asset ì •ë³´ê°€ ìˆëŠ” ìµœìƒìœ„ í‚¤ë¥¼ ìë™ìœ¼ë¡œ ì°¾ê¸°
    asset_item = asset_resp.get("assetItem") or asset_resp.get("asset") or asset_resp

    type_id = asset_item.get("typeIdentifier")
    metadata = asset_item  # ì „ì²´ ë‚´ìš©ì„ metadataë¡œ ì €ì¥

    # schemaê°€ ì¡´ì¬í•˜ë©´ ê°€ì ¸ì˜¤ê¸° (Asset íƒ€ì…ì— ë”°ë¼ í•„ë“œëª… ë‹¤ë¥¼ ìˆ˜ ìˆìŒ)
    schema = asset_item.get("schema") or asset_item.get("glueTable", {}).get("columns")

    # 5ï¸âƒ£ JSON ì €ì¥ìš© ë°ì´í„° ì •ë¦¬
    asset_data = {
        "identifier": asset_id,
        "name": asset_name,
        "typeIdentifier": type_id,
        "metadata": metadata,
        "schema": schema
    }

    all_assets_data.append(asset_data)

# 6ï¸âƒ£ JSON íŒŒì¼ë¡œ ì €ì¥
output_file = "sample_result_data/datazone_assets_detailed.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(all_assets_data, f, default=str, ensure_ascii=False, indent=2)

print(f"âœ… Detailed asset info saved to '{output_file}'")
````

### ì•„ë˜ì˜ ì½”ë“œëŠ” ì „ì²´ Asset ì¤‘ GlueTableAssetTypeë§Œ ì°¾ì•„ì„œ í•´ë‹¹ Asset ë³„ ë©”íƒ€ë°ì´í„°ë¥¼ "[Assetëª…]_[AssetID].json"í˜•ì‹ìœ¼ë¡œ ì €ì¥í•˜ëŠ” ì˜ˆì‹œ 

````python
import boto3
import json
import os
import re

region = "ap-northeast-2"
dz = boto3.client("datazone", region_name=region)
domain_id = "dzd-cjvglgj4d43fmg"

# ê²°ê³¼ ì €ì¥ í´ë”
output_dir = "sample_result_data/glue_assets"
os.makedirs(output_dir, exist_ok=True)

# âœ… 1ï¸âƒ£ ë„ë©”ì¸ ë‚´ ëª¨ë“  í”„ë¡œì íŠ¸ ì¡°íšŒ
projects_resp = dz.list_projects(domainIdentifier=domain_id)
projects = {p["id"]: p["name"] for p in projects_resp.get("items", [])}

print(f"ğŸ” Found {len(projects)} projects in domain {domain_id}\n")

# âœ… 2ï¸âƒ£ ëª¨ë“  í”„ë¡œì íŠ¸ë³„ GlueTableAssetType ì¡°íšŒ
glue_assets = []

for project_id, project_name in projects.items():
    print(f"ğŸ“ Searching assets in project: {project_name} ({project_id})")

    next_token = None
    while True:
        params = {
            "domainIdentifier": domain_id,
            "owningProjectIdentifier": project_id,
            "searchScope": "ASSET"
        }
        if next_token:
            params["nextToken"] = next_token

        resp = dz.search(**params)

        for item in resp.get("items", []):
            asset_item = item.get("assetItem", {})
            if asset_item.get("typeIdentifier") == "amazon.datazone.GlueTableAssetType":
                glue_assets.append({
                    "id": asset_item.get("identifier"),
                    "name": asset_item.get("name"),
                    "type": asset_item.get("typeIdentifier"),
                    "project_id": project_id,
                    "project_name": project_name,
                })

        next_token = resp.get("nextToken")
        if not next_token:
            break

print(f"\nâœ… Found total {len(glue_assets)} GlueTableAssetType assets.\n")

# âœ… 3ï¸âƒ£ ê° Assetì— ëŒ€í•´ get_asset í˜¸ì¶œ í›„ ê°œë³„ JSON ì €ì¥
for asset in glue_assets:
    asset_id = asset["id"]
    asset_name = asset["name"]

    # íŒŒì¼ëª…ì— ì“¸ ìˆ˜ ì—†ëŠ” ë¬¸ì ì œê±° (ì˜ˆ: /, :, ?, ë“±)
    safe_name = re.sub(r'[^a-zA-Z0-9ê°€-í£_\-]', '_', asset_name)

    output_path = os.path.join(output_dir, f"{safe_name}_{asset_id}.json")

    print(f"ğŸ“¦ Fetching asset: {asset_name} ({asset_id})")

    try:
        # Asset ìƒì„¸ ì¡°íšŒ
        asset_resp = dz.get_asset(
            domainIdentifier=domain_id,
            identifier=asset_id
        )

        asset_item = asset_resp.get("assetItem") or asset_resp.get("asset") or asset_resp
        type_id = asset_item.get("typeIdentifier")
        metadata = asset_item
        schema = asset_item.get("schema") or asset_item.get("glueTable", {}).get("columns")

        # ì €ì¥ ë°ì´í„° êµ¬ì¡°
        asset_data = {
            "identifier": asset_id,
            "name": asset_name,
            "project": asset["project_name"],
            "typeIdentifier": type_id,
            "metadata": metadata,
            "schema": schema
        }

        # JSON íŒŒì¼ë¡œ ì €ì¥
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(asset_data, f, default=str, ensure_ascii=False, indent=2)

        print(f"âœ… Saved: {output_path}")

    except Exception as e:
        print(f"âŒ Failed to fetch asset {asset_name} ({asset_id}): {e}")

print("\nğŸ‰ All GlueTableAssetType asset metadata saved successfully!")
````
